```python
import pandas as pd

dataset = pd.read_csv('AB_NYC_2019.csv')
dataset.head()
```

```python
dataset.info()

dataset.head()
# The dataset has 48,895 rows and 16 columns. The columns can be categorized into different types:

#     Numerical: 'id', 'host_id', 'latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365'
#     Categorical: 'name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type'
#     Date: 'last_review'

# Some columns like 'name', 'host_name', 'last_review', and 'reviews_per_month' have missing values.

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


fig, axes = plt.subplots(3, 2, figsize=(14, 10))

axes = axes.flatten()



# Specify columns to apply log transformation

log_transform_cols = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count']
numerical_cols = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
summary_stats = dataset[numerical_cols].describe()


for i, col in enumerate(numerical_cols):

    ax = axes[i]

    data = dataset[col]  # Using 'dataset' instead of 'df'

    

    # Apply log transformation if column is in the list

    if col in log_transform_cols:

        # Adding a small constant to handle zero values

        data = np.log1p(data)

        ax.set_title(f'Log-transformed Distribution of {col}')

    else:

        ax.set_title(f'Distribution of {col}')

    

    sns.histplot(data, bins=30, kde=True, ax=ax)

    ax.set_xlabel(col)

    ax.set_ylabel('Frequency')



plt.tight_layout()

plt.show()



summary_stats
```

```python
# List of categorical columns
categorical_cols = ['name', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type']

# Count unique values and their frequencies
unique_counts = {}
for col in categorical_cols:
    unique_counts[col] = dataset[col].nunique()

# Generate bar plots for categorical columns with fewer unique values
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes = axes.flatten()

# Selecting columns with fewer unique values for visualization
selected_cols = ['neighbourhood_group', 'room_type']

for i, col in enumerate(selected_cols):
    sns.countplot(data=dataset, x=col, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Frequency')

# Date field 'last_review' as a time series plot
dataset['last_review'] = pd.to_datetime(dataset['last_review'])
dataset['last_review'].dropna().dt.to_period("M").value_counts().sort_index().plot(kind='line', ax=axes[2])
axes[2].set_title('Time Series of Last Review Dates')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

unique_counts
```

```python
correlation_matrix = dataset[numerical_cols].corr()



# Create a heatmap to visualize the correlations

plt.figure(figsize=(10, 8))

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')

plt.title("Correlation Matrix of Numerical Variables")

plt.show()
```

```python
# Create box plots to visualize the influence of 'Room Type' and 'Neighbourhood Group' on 'Price'
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Box plot for 'Room Type' vs 'Price'
sns.boxplot(data=dataset, x='room_type', y='price', ax=axes[0])
axes[0].set_title('Price Distribution by Room Type')
axes[0].set_yscale('log')  # Applying log scale due to wide range of prices
axes[0].set_ylabel('Price (Log Scale)')

# Box plot for 'Neighbourhood Group' vs 'Price'
sns.boxplot(data=dataset, x='neighbourhood_group', y='price', ax=axes[1])
axes[1].set_title('Price Distribution by Neighbourhood Group')
axes[1].set_yscale('log')  # Applying log scale due to wide range of prices
axes[1].set_ylabel('Price (Log Scale)')

plt.tight_layout()
plt.show()
```

```python
# Assuming 'dataset' is your DataFrame and 'last_review' is converted to datetime format
dataset['last_review'] = pd.to_datetime(dataset['last_review'])

# 1. Days Since Last Review
# Assuming you have a reference date, perhaps the maximum date in the 'last_review' column
reference_date = dataset['last_review'].max()
dataset['days_since_last_review'] = (reference_date - dataset['last_review']).dt.days

# 2. Is Superhost
# Creating a binary feature based on certain metrics (e.g., high number of reviews and high review frequency)
high_review_count_threshold = dataset['number_of_reviews'].quantile(0.75)
high_review_frequency_threshold = dataset['review_frequency'].quantile(0.75)
dataset['is_superhost'] = ((dataset['number_of_reviews'] >= high_review_count_threshold) & 
                           (dataset['review_frequency'] >= high_review_frequency_threshold)).astype(int)

# 3. Review-to-Availability Ratio
# Calculating the ratio of 'Number of Reviews' to 'Availability 365'
dataset['review_to_availability_ratio'] = dataset['number_of_reviews'] / dataset['availability_365']

# Handle division by zero and infinite values by replacing them with NaN
dataset.replace([np.inf, -np.inf], np.nan, inplace=True)

# Display the dataset to confirm that the new features are added
print(dataset[['days_since_last_review', 'is_superhost', 'review_to_availability_ratio']].head())

# Calculates the Days Since Last Review by finding the difference between a reference date (the latest review date in the dataset) and the 'last_review' for each listing.
# Identifies potential Superhosts based on being in the 75th percentile or higher for both the number of reviews and review frequency.
# Computes the Review-to-Availability Ratio by dividing the number of reviews by the availability over 365 days.
```

```python
# Calculate the new features in the dataset

# Days Since Last Review

reference_date = dataset['last_review'].max()

dataset['days_since_last_review'] = (reference_date - dataset['last_review']).dt.days



# Is Superhost

high_review_count_threshold = dataset['number_of_reviews'].quantile(0.75)

high_review_frequency_threshold = dataset['review_frequency'].quantile(0.75)

dataset['is_superhost'] = ((dataset['number_of_reviews'] >= high_review_count_threshold) & 

                           (dataset['review_frequency'] >= high_review_frequency_threshold)).astype(int)



# Review-to-Availability Ratio

dataset['review_to_availability_ratio'] = dataset['number_of_reviews'] / dataset['availability_365']



# Handle division by zero and infinite values by replacing them with NaN

dataset.replace([np.inf, -np.inf], np.nan, inplace=True)



# Generate summary statistics for the new features

new_features = ['days_since_last_review', 'is_superhost', 'review_to_availability_ratio']

summary_stats_new_features = dataset[new_features].describe()



# Plotting histograms for the new features

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes = axes.flatten()



for i, col in enumerate(new_features):

    if col != 'is_superhost':  # Histogram doesn't make sense for a binary feature

        sns.histplot(dataset[col], bins=30, kde=False, ax=axes[i])

    else:

        sns.countplot(data=dataset, x=col, ax=axes[i])

    axes[i].set_title(f'Distribution of {col}')

    axes[i].set_xlabel(col)

    axes[i].set_ylabel('Frequency')



plt.tight_layout()

plt.show()



summary_stats_new_features
```

```python
extended_numerical_cols = numerical_cols + new_features



# Generate an extended correlation matrix that includes the new features

extended_correlation_matrix = dataset[extended_numerical_cols].corr()



# Create a heatmap to visualize the extended correlations

plt.figure(figsize=(12, 10))

sns.heatmap(extended_correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')

plt.title("Extended Correlation Matrix Including Engineered Features")

plt.show()
```
